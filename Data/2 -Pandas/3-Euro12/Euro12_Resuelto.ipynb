### Step 3. Assign it to a variable called euro12.

```python
discipline = euro12[["Team","Yellow Cards","Red Cards"]]
discipline
```

### Step 8. Sort the teams by Red Cards, then to Yellow Cards

```python
red = euro12.loc[:,"Red Cards"].sort_values(ascending=False)
```

```python
euro12.loc[red,"Team"]
```

```python
yellow = euro12.loc[:,"Yellow Cards"].sort_values(ascending=False)
```

```python
euro12.loc[yellow,"Team"]
```

### Step 9. Calculate the mean Yellow Cards given per Team

```python
sum(euro12.loc[:,"Yellow Cards"])/len(euro12.loc[:,"Team"])
```
