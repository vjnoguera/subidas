## 1. Create an interactive bar chart showing total sales by State for2014.

```python
sp1 = sp.loc[sp.loc[:,"Order Date"].dt.year==2014]
sp1.groupby("State")[["Sales"]].sum().reset_index()
```

```python
# prepare data frames
sp1_1 = sp1.groupby("State")[["Sales"]].sum().reset_index()
# import graph objects as "go"
# import plotly.graph_objects as go

# create trace1 
trace1 = go.Bar(x = sp1_1['State'],
               y = sp1_1['Sales'],
               name = 'Sales Relation',
               marker = dict(color = 'rgba(255, 174, 255, 0.5)',
                            line = dict(color='rgb(0,0,0)', width = 1.5)),
               text = sp1_1['State'])

# trace2 = go.Bar(x = df2014['university_name'],
#                y = df2014['teaching'],
#                name = 'teaching',
#                marker = dict(color = 'rgba(255, 255, 128, 0.5)',
#                             line = dict(color='rgb(0,0,0)', width = 1.5)),
#                text = df2014['country'])



data = [trace1]

layout = go.Layout(barmode = "group")

fig = go.Figure(data = data, layout = layout)

fig
```

## 2. Create an interactive line chart showing Sales and Profit in 2014.

```python
sp1 = sp1.sort_values("Order Date")
sp1

```

```python
# prepare data frame
# df = timesData.iloc[:100,:]

# import graph objects as "go"
# import plotly.graph_objects as go

# Creating trace1
trace1 = go.Scatter(
                    x = sp1.loc[:,"Order Date"].sort_values(ascending=True)
,
                    y = sp1['Sales'],
                    name = 'Sales',# es importante siempre ponerle un nombre a la traza
                    mode= 'lines',
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),# el color siempre se introduce como diccionario, y podemos utilizar el rgba o poner el nombre "green"
                    text = sp1['Sales'])

# Creating trace2
trace2 = go.Scatter(
                    x = sp1.loc[:,"Order Date"].sort_values(ascending=True)
,
                    y = sp1['Profit'],
                    name = 'Profits',
                    mode= 'lines',
                    marker = dict(color = 'rgba(80, 26, 80, 0.8)'),
                    text = sp1['Profit'])

''' el gr√°fico siempre tendra, data y layout'''
data = [trace1, trace2]

layout = dict(title = 'Sales and Profit in 2014', # existe un objeto go.layout que requiere los mismos elemento y puede ser incluso mas interesante
             xaxis= dict(title= 'World Rank',ticklen= 5)
           )

fig = go.Figure(data = data, layout=layout)

fig
```

## 4. Create an interactive stacked bar chart showing the revenue by state (bars) for each of the products categories for California, New York and Texas.

```python
sp4 = sp.loc[sp.loc[:,"State"].isin(["Texas","New York","California"])]
sp4_1 = sp4.groupby(["State","Category"],as_index=False)[["Sales"]].sum()# es imprescindible meterse en el as_index=False para eliminar el multiindice que se nos generar√≠a
sp4_1
```

```python
import plotly.express as px
fig = px.bar(sp4_1, x="Category", y = "Sales", color = "State")
fig.show()

```

```python
sp4_1 = sp4.groupby(["State","Category"]).size().reset_index().pivot(columns='State',
                                                                                  index='Category',
                                                                                  values=0)
sp4_1.plot(kind='bar', stacked=True);
```
